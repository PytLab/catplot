# 绘制三维平面

```python
%matplotlib inline
```

## 创建一个三维平面

```python
from catplot.grid_components.planes import Plane3D
```

```python
plane = Plane3D([1, 3], [1, 3], 3, color="#CD4125", edgecolor="", alpha=0.5, shade=True)
```

## 创建三维画布

```python
from catplot.grid_components.grid_canvas import Grid3DCanvas
```

```python
canvas = Grid3DCanvas()
```

## 绘制

```python
canvas.add_plane(plane)
```

```python
canvas.draw()
canvas.figure
```

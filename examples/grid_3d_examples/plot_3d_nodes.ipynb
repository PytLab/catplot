# 绘制三维nodes

## 创建3D Nodes

```python
from catplot.grid_components.nodes import Node3D
```

```python
node0 = Node3D([0.0, 0.0, 0.0], color="#D43D37")
node1 = Node3D([0.5, 0.5, 0.5], color="#2180CC")
node2 = Node3D([1.0, 1.0, 1.0], color="#AC7E02")
node3 = Node3D([0.5, 0.7, 0.8])
```

## 创建3D 画布

```python
from catplot.grid_components.grid_canvas import Grid3DCanvas
```

```python
canvas = Grid3DCanvas()
```

## 将创建的nodes添加到三维画布中

```python
canvas.add_nodes([node0, node1, node2, node3])
```

## 绘制

```python
canvas.draw()
```

```python
canvas.figure
```

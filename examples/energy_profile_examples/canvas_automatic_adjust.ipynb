```python
%matplotlib inline
from catplot.ep_components.ep_canvas import EPCanvas
from catplot.ep_components.ep_lines import ElementaryLine
```

# 画布会随着内容自动更新大小

```python
canvas = EPCanvas()
```

```python
line = ElementaryLine([0.0, 1.3, 0.8])
```

```python
canvas.add_line(line)
canvas.draw()
canvas.figure
```

```python
line.translate_state("FS", -1.0)
```

```python
canvas.draw()
```

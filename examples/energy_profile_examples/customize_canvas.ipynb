# 画布定制

```python
from catplot.ep_components.ep_canvas import EPCanvas
```

## 创建自定义画布

```python
canvas = EPCanvas(margin_ratio=0.2,     # 边缘留白的比例
                  figsize=(6, 4),       # 画布大小(单位：英寸)
                  dpi=200,              # 每英寸所包含的像素点
                  facecolor="#C5C5C5",  # 画布背景色
                  edgecolor="#CE594B")  # 边框的颜色
```

> 可以看到上面的画布大小、边框颜色以及背景色与默认均不相同
